```python
import os
import pandas as pd
import pickle as pkl
```

```python
results_folder='/nfs/iil/home/gkoren2/work/rad_data/results/'
os.listdir(results_folder)
```

```python
lsdf=pd.read_csv(os.path.join(results_folder,'lab_scores.csv'))
lsdf
```

```python
lsdf2=lsdf.rename(columns={' seed':'seed'})
lsdf2.set_index(['model type','seed'],inplace=True)
lsdf2.sort_index(inplace=True)
lsdf2
```

```python
def analyze_res_dict(res,gt):
    spcorr={'img-enc':None,'fc1':None,'relu1':None}
    res_dict={'model type':[],'seed':[],'emb':[],'score':[]}
    for k,v in res.items():
        emb=k.split('_')[0]
        ms=k[k.find('dr'):]
        m=ms.split('-seed')[0]
        s=ms.split('-seed')[1]
        res_dict['model type'].append(m)
        res_dict['seed'].append(int(s))
        res_dict['emb'].append(emb)
        res_dict['score'].append(v)
    res_df=pd.DataFrame(res_dict)
    res_df2=res_df.pivot(index=['model type','seed'],columns=['emb'],values=['score'])
    res_df2.sort_index(inplace=True)
    gtdf=gt.join(res_df2)
    spcorr['img-enc']=gtdf.loc[:,[' mean rewards',('score', 'img-encoder')]].corr(method='spearman').values[0,1]
    spcorr['fc1']=gtdf.loc[:,[' mean rewards',('score', 'fc1')]].corr(method='spearman').values[0,1]
    spcorr['relu1']=gtdf.loc[:,[' mean rewards',('score', 'relu1')]].corr(method='spearman').values[0,1]
    return spcorr
```

# prepare summary dataframe

## res_s

```python
res_s_files=[f for f in os.listdir(results_folder) if 'vs.pkl' in f]
print(len(res_s_files))
res_s_files
```

```python
res_s=dict()
for f in res_s_files:
    fname=os.path.join(results_folder,f)
    with open(fname,'rb') as fp:
        res_s.update(pkl.load(fp))
print(len(res_s))
res_s
```

```python
analyze_res_dict(res_s,lsdf2)
```
